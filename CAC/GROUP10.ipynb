```python
import csv
import matplotlib.pyplot as plt

# Initialize variables to store gender data
gender_counts = {"Male": 0, "Female": 0, "Other": 0}  # Assuming "Other" for any additional genders

# Open and read the CSV file
with open('Final.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        gender = row['Gender']
        if gender in gender_counts:
            gender_counts[gender] += 1
        else:
            gender_counts["Other"] += 1

# Extract the data for the pie chart
labels = list(gender_counts.keys())
sizes = list(gender_counts.values())

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Gender Distribution')

# Display the chart
plt.show()
```

```python
#======Course========
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
df = pd.read_csv('Final.csv')

# Create a count plot of the "Course" column
plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
sns.set(style="whitegrid")  # Optional: Set a style for the plot
sns.countplot(data=df, x='Which course are you studying in?')

# Add labels and title
plt.xlabel('Course')
plt.ylabel('Count')
plt.title('Count Plot of Courses')

# Display the plot
plt.xticks(rotation=90)  # Optional: Rotate x-axis labels for better readability
plt.show() 
```

```python
#=========Current Year============
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re

# Load the CSV data into a DataFrame
df = pd.read_csv('Final.csv')

# Extract numerical values from the "Current Year" column
df['What is your current year of study ?'] = df['What is your current year of study ?'].apply(lambda x: int(re.search(r'\d+', x).group()))

# Create a distplot of the "Current Year" column
plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
sns.set(style="whitegrid")  # Optional: Set a style for the plot
sns.distplot(df['What is your current year of study ?'], bins=10, kde=True, hist=True)

# Add labels and title
plt.xlabel('Current Year')
plt.ylabel('Density')
plt.title('Distribution Plot of Current Year')

# Display the plot
plt.show()
```

```python
#===============Peak time==================
import csv
import matplotlib.pyplot as plt

# Initialize variables to store peak hours data
peak_hours_counts = {"Lunch Time": 0, "After class hours": 0, "Break time": 0, "Other": 0}  # Assuming "Other" for any additional values
explode = (0, 0, 0.1, 0)  # Explode the "Break time" slice

# Open and read the CSV file
with open('Final.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        peak_hour = row['What are the peak hours in the trinity store when students tend to shop the most?']
        if peak_hour in peak_hours_counts:
            peak_hours_counts[peak_hour] += 1
        else:
            peak_hours_counts["Other"] += 1

# Extract the data for the pie chart
labels = list(peak_hours_counts.keys())
sizes = list(peak_hours_counts.values())

# Create a pie chart with an exploded slice
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, explode=explode)

# Add a title
plt.title('Peak Hours Distribution')

# Display the chart
plt.show()
```

```python
#===============Peak time==================
import csv
import matplotlib.pyplot as plt

# Initialize variables to store peak hours data
year_counts = {"1st year": 0, "2nd year": 0, "3rd year": 0, "4th year": 0, "5th year":0}  # Assuming "Other" for any additional values
explode = (0.1, 0, 0, 0, 0)  # Explode the "Break time" slice

# Open and read the CSV file
with open('Final.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        year_hour = row['What is your current year of study ?']
        if year_hour in year_counts:
            year_counts[year_hour] += 1
        else:
            year_counts["Other"] += 1

# Extract the data for the pie chart
labels = list(year_counts.keys())
sizes = list(year_counts.values())

# Create a pie chart with an exploded slice
plt.pie(sizes, labels=labels, autopct='%1.0f%%', startangle=500, explode=explode)

# Add a title
plt.title('Year Distribution')

# Display the chart
plt.show()
```
