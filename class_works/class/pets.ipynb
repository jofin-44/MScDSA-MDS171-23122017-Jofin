```python
class PetStore:
    def _init_(self):
        self.pets = {}

    def store_pet(self, name, species, price):
        self.pets[name] = {'species': species, 'price': price}

    def search_pet(self, name):
        if name in self.pets:
            return self.pets[name]
        else:
            return "Pet not found."

    def sell_pet(self, name):
        if name in self.pets:
            del self.pets[name]
            return "Pet sold."
        else:
            return "Pet not found."

    def list_pets(self):
        return self.pets

# Sample usage of the PetStore class
if __name__ == "_main_":
    pet_store = PetStore()

    pet_store.store_pet("Fluffy", "Cat", 50)
    pet_store.store_pet("Buddy", "Dog", 100)
    pet_store.store_pet("Squawk", "Parrot", 30)

    while True:
        print("\nMenu:")
        print("1. List all pets")
        print("2. Search for a pet")
        print("3. Sell a pet")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            print("List of all pets:")
            print(pet_store.list_pets())

        elif choice == '2':
            pet_name = input("Enter the pet name to search: ")
            result = pet_store.search_pet(pet_name)
            print(result)

        elif choice == '3':
            pet_name = input("Enter the pet name to sell: ")
            result = pet_store.sell_pet(pet_name)
            print(result)

        elif choice == '4':
            print("Exiting the program.")
            break
```
