Write a python program with following specification ,there will be a function named create dataset that takes dictionary,file name ,No. of rows argument
Dictionary has the following structure:
{
    "column_name" : [domain of values]
}
your function will select random values for each of the domains and create a row for you. The number of row that has to be created is the parameter that has been taken.
Save the file with required no. of rows.
Load this data set using pandas and display basic information         

```python
import pandas as pd
import random
dict={
        "Name" : ["Arun","Anand","Hiran","John","Abhidev","Aleena","Nanadana","George"],
        "Age" : [20,68,28,30,12,13,24],
        "Gender" : ["Male","Male","Male","Male","Male","Female","Female","Male"],
        "Place" : ["Palakkad","Thrissur","Thrissur","Thrissur","Trivandrum","Kannur","Thrissur","Kottayam"],
        "Weight" : [52,98,78,56,43,14,34,39],
        "Height" : [167,164,163,178,175,170,176,166]
        }

rows = int(input("Enter the number of rows required:"))
file_name = input("Enter the file name:")
file_name = file_name +".csv"

def create_dataset(dict,file_name,rows):
    file = open(file_name,"w")
    file.write("")
    file.close()

    file = open(file_name,"a")
    file.write("Name"+","+"Age"+","+"Gender"+","+"Place"+","+"Weight"+","+"Height"+"\n")
    for i in range(0,rows,1):
        name = random.choice(dict["Name"])
        age = random.choice(dict["Age"])
        gender = random.choice(dict["Gender"])
        place = random.choice(dict["Place"])
        weight = random.choice(dict["Weight"])
        height = random.choice(dict["Height"])
        file.write(name + "," + str(age) + "," + gender + "," + place + "," + str(weight) + "," + str(height) + "\n" )
    file.close()
create_dataset(dict,file_name,rows)

```
