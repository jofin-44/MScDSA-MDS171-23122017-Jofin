```python
import random
def create_order():
    order_data = []
    fd_list = ["pasta", "pizza", "noodles", "burger", "sandwich"]
    while True:
        print("1.Create a new order")
        print("2.View orders")
        print("3.Exit")
        choice = int(input("Enter the choice:"))

        if choice == 1:
            order = {}
            r1 = random.randint(0, 9)
            r2 = random.randint(0, 9)
            r3 = random.randint(0, 9)
            r4 = random.randint(0, 9)
            r5 = random.randint(0, 9)
            order_id = str(r1) + str(r2) + str(r3) + str(r4) + str(r5)
            order["order_id"] = order_id
            items = []

            while True:
                print("1.Pasta")
                print("2.Pizza")
                print("3.Noodles")
                print("4.Burger")
                print("5.Sandwich")
                print("6.Finish Order")
                item_choice = int(input("Enter the item needed:"))

                if item_choice == 1:
                    item = {}
                    item["item"] = fd_list[0]
                    item["item number"] = int(input("Item Number:"))
                    item["quantity"] = int(input("Quantity:"))
                    items.append(item)
                elif item_choice == 2:
                    item = {}
                    item["item"] = fd_list[1]
                    item["item number"] = int(input("Item Number:"))
                    item["quantity"] = int(input("Quantity:"))
                    items.append(item)
                elif item_choice == 3:
                    item = {}
                    item["item"] = fd_list[2]
                    item["item number"] = int(input("Item Number:"))
                    item["quantity"] = int(input("Quantity:"))
                    items.append(item)
                elif item_choice == 4:
                    item = {}
                    item["item"] = fd_list[3]
                    item["item number"] = int(input("Item Number:"))
                    item["quantity"] = int(input("Quantity:"))
                    items.append(item)
                elif item_choice == 5:
                    item = {}
                    item["item"] = fd_list[4]
                    item["item number"] = int(input("Item Number:"))
                    item["quantity"] = int(input("Quantity:"))
                    items.append(item)
                elif item_choice == 6:
                    order["items"] = items
                    order_data.append(order)
                    break

        elif choice == 2:
            print("The orders are :")
            for order in order_data:
                print(order)
            with open("orders_list","a") as file:
                for order in order_data:
                    file.write(str(order)+"\n")
        elif choice == 3:
            print("=======E X I T E D========")
            break
create_order()
```
